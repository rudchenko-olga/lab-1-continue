```python
#Доступ по индексу в DataFrame
```

```python
 df = pd.DataFrame({
... 'country': ['Belgium', 'France', 'Belarus', 'Ukraine'],
... 'population': [17.04, 143.5, 9.5, 45.5],
... 'square': [2724902, 17125191, 207600, 603628]
... }, index=['BE', 'FR', 'BY', 'UA'])
```

```python
df
```

```python
df.index = ['BE', 'FR', 'BY', 'UA']
```

```python
df.index.name = 'Country Code'
```

```python
df
```

```python
# индексу было задано имя - Country Code. объекты Series из DataFrame будут иметь те же индексы, что и объект DataFrame:
```

```python
df['country']
```

```python
#Доступ к строкам по индексу возможен несколькими способами:
# .loc - используется для доступа по строковой метке
# .iloc - используется для доступа по числовому значению (начиная от 0)
```

```python
df.loc ['BE']
```

```python
df.iloc[0]
```

```python
#выборка по индексу и интересующим колонкам:
```

```python
df.loc[['BE', 'FR'], 'population']
```

```python
 df.loc['BE':'BY', :]
```

```python
#Фильтровать DataFrame с помощью т.н. булевых массивов:
```

```python
df[df.population > 10][['country', 'square']]
```

```python
#Сбросить индексы 
```

```python
df.reset_index()
```

```python
#Добавим новый столбец, в котором население (в миллионах) поделим наплощадь страны, получив тем самым плотность:
```

```python
df['density'] = df['population'] / df['square'] * 1000000
```

```python
df
```

```python
#удалить новый столбец
```

```python
df.drop(['density'], axis='columns')
```

```python
#Переименовывать столбцы нужно через метод rename:
```

```python
df = df.rename(columns={'Country Code': 'country_code'})
```
